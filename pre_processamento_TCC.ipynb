## Conversão "Label" em numeral

```
# Conversão "Label" em numeral
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
UDPlag[' Label'] = labelencoder.fit_transform(UDPlag[' Label'])
syn[' Label'] = labelencoder.fit_transform(syn[' Label'])
UDP[' Label'] = labelencoder.fit_transform(UDP[' Label'])
LDAP[' Label'] = labelencoder.fit_transform(LDAP[' Label'])
msql[' Label'] = labelencoder.fit_transform(msql[' Label'])
netbios[' Label'] = labelencoder.fit_transform(netbios[' Label'])

print('Ataque de exploração UDPLag:', UDPlag.groupby(by=' Label').size(), 
      '\n\nAtaque de exploração Syn:', syn.groupby(by=' Label').size(),
      '\n\nAtaque de exploração UDP:', UDP.groupby(by=' Label').size(),
      '\n\nAtaque de exploração LDAP:', LDAP.groupby(by=' Label').size(),
      '\n\nAtaque de exploração MSSQL:', msql.groupby(by=' Label').size(),
      '\n\nAtaque de exploração NetBIOS:', netbios.groupby(by=' Label').size())
```

## Remoção das Informações do soquete

```
# Remoção das Informações do soquete
UDPlag.drop(['Unnamed: 0' ,'Flow ID', ' Source IP', ' Source Port',
       ' Destination IP', ' Destination Port', ' Protocol', ' Timestamp', 'SimillarHTTP', ' Inbound'], axis=1, inplace=True)
syn.drop(['Unnamed: 0' ,'Flow ID', ' Source IP', ' Source Port',
       ' Destination IP', ' Destination Port', ' Protocol', ' Timestamp', 'SimillarHTTP', ' Inbound'], axis=1, inplace=True)
UDP.drop(['Unnamed: 0' ,'Flow ID', ' Source IP', ' Source Port',
       ' Destination IP', ' Destination Port', ' Protocol', ' Timestamp', 'SimillarHTTP', ' Inbound'], axis=1, inplace=True)
LDAP.drop(['Unnamed: 0' ,'Flow ID', ' Source IP', ' Source Port',
       ' Destination IP', ' Destination Port', ' Protocol', ' Timestamp', 'SimillarHTTP', ' Inbound'], axis=1, inplace=True)
msql.drop(['Unnamed: 0', 'Flow ID', ' Source IP', ' Source Port',
       ' Destination IP', ' Destination Port', ' Protocol', ' Timestamp', 'SimillarHTTP', ' Inbound'], axis=1, inplace=True)
netbios.drop(['Unnamed: 0', 'Flow ID', ' Source IP', ' Source Port',
       ' Destination IP', ' Destination Port', ' Protocol', ' Timestamp', 'SimillarHTTP', ' Inbound'], axis=1, inplace=True)
```

```
UDPlag.head()
```

```
#Remoção dos espaços em branco
def correct_space(space):
    space = space.replace(' ', '')
    return space
```

```
UDPlag = UDPlag.apply(correct_space)
syn = syn.apply(correct_space)
UDP = UDP.apply(correct_space)
LDAP = LDAP.apply(correct_space)
msql = msql.apply(correct_space)
netbios = netbios.apply(correct_space)
```

## Remoção dos valores infinite e valores NaN



```
UDPlag.replace([np.inf, -np.inf], np.nan, inplace=True)
syn.replace([np.inf, -np.inf], np.nan, inplace=True)
UDP.replace([np.inf, -np.inf], np.nan, inplace=True)
LDAP.replace([np.inf, -np.inf], np.nan, inplace=True)
msql.replace([np.inf, -np.inf], np.nan, inplace=True)
netbios.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```
UDPlag = UDPlag.dropna()
syn = syn.dropna()
UDP = UDP.dropna()
LDAP = LDAP.dropna()
msql = msql.dropna()
netbios = netbios.dropna()
```
